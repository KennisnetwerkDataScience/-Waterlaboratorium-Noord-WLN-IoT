```python
import pandas as pd
import xlwings as xw
pd.set_option('display.max_columns',200)
```

```python
PATH = 'C:/Users/kleis500/Eclipse workspace/KNN_DS/Waterlaboratorium-Noord-IoT/Data/Kennisnetwerk_data_WWTPpopulaties_WLN.xlsx'
wb = xw.Book(PATH)
sheet = wb.sheets['Pivot_ZAWZI_Ranked']

df1 = sheet['A1:AP2000'].options(pd.DataFrame, index=False, header=False).value
```

```python
df1.head(15)
```

```python
df1_part1 = df1.loc[9:,0:9].copy()
df1_part1.columns = df1_part1.iloc[0]
df1_part1.drop(labels = 9, axis = 0, inplace = True)
df1_part1.head()
```

```python
df1_part2 = df1.loc[4:,9:].copy()
df1_part2.drop(labels = [4,6,7,8, 9], axis = 0, inplace = True)
df1_part2.head()
```

```python
df1_part2.columns = df1_part2.iloc[0]
df1_part2.drop(5, axis = 0, inplace = True)
df1_part2 = df1_part2.rename(columns={"Sample": "Difference"}).copy()
# df1_part2.head()
```

```python
df_bac = df1_part1.merge(df1_part2, how='inner', on='Difference')
df_bac.head()
```

```python
df_selected = df_bac.loc[df_bac.Difference >= 2.00]
df_selected.Class.unique()
print('This df contains:', len(df_selected.Class.unique()),'different classes of bacteria\nand has the shape', df_selected.shape)
```
